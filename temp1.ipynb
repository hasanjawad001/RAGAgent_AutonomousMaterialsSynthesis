```python
import pickle
import networkx as nx
import os

# Define the file path
file_path = os.path.join("outputs", "d1a.pkl")

# 1. Load the graph object from the pickle file
try:
    with open(file_path, 'rb') as f:
        graph = pickle.load(f)
    print(f"‚úÖ Successfully loaded graph from: {file_path}")
except FileNotFoundError:
    print(f"‚ö†Ô∏è Error: File not found at {file_path}. Check your path!")
    exit()
except Exception as e:
    print(f"‚ö†Ô∏è Error loading graph: {e}")
    exit()
```

```python
# 2. Check the type of the loaded object
print(f"Type of the loaded object: {type(graph)}")

# 3. Inspect the graph (Assuming it's a NetworkX Graph or a compatible object)
if isinstance(graph, nx.Graph) or isinstance(graph, nx.DiGraph):
    print("\n--- Graph Summary ---")
    print(f"Total Nodes: {graph.number_of_nodes()}")
    print(f"Total Edges: {graph.number_of_edges()}")
    
    # Check the first few nodes and their attributes
    print("\nLast 10 Nodes (ID and Attributes):")
    for i, (node_id, data) in enumerate(graph.nodes(data=True)):
        if i >= graph.number_of_nodes() - 10:
            print(f"  - Node ID: {node_id}, Attributes: {data}")

    # Check the first few edges and their attributes
    print("\nLast 10 Edges (Source, Target, and Attributes):")
    for i, (source, target, data) in enumerate(graph.edges(data=True)):
        if i >= graph.number_of_edges() - 10:
            print(f"  - Edge: ({source}) -> ({target}), Attributes: {data}")
else:
    print(f"\nüí° Note: The loaded object is not a standard NetworkX graph type.")
    print("You may need to cast it or access its internal graph attribute (e.g., graph.graph).")
```
